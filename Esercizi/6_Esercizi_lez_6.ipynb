### **1. Distribuzioni di probabilità**
Genera 1000 valori casuali da una distribuzione gamma con parametro di forma pari a 1.
Suggerimento: il parametro di forma viene passato come primo argomento quando si "congela" la distribuzione.

Traccia l’istogramma del campione e sovrapponi la PDF della distribuzione.

Stima il parametro di forma dal campione usando il metodo fit.

Extra:
Le distribuzioni hanno molti metodi utili. Esplorali usando il completamento automatico con il tasto TAB.

Traccia la funzione di distribuzione cumulativa (CDF).

Calcola la varianza.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp

# Congelo la distribuzione gamma con parametro di forma a=1
gamma_dist = sp.stats.gamma(a=1)

# 1000 valori casuali
sample = gamma_dist.rvs(size=1000)

# Istogramma
plt.hist(sample, bins=50, density=True, alpha=0.6, label='Istogramma campione')

# PDF teorica 
x = np.linspace(0, np.max(sample))
plt.plot(x, gamma_dist.pdf(x), label='PDF teorica')

plt.title("Distribuzione Gamma (a=1)")
plt.xlabel("x")
plt.ylabel("Densità")
plt.legend()
plt.tight_layout()
plt.show()

# Stima dei parametri della distribuzione Gamma dal campione
a_hat, loc_hat, scale_hat = sp.stats.gamma.fit(sample)

print(f"Parametro stimato (a): {a_hat}")
print(f"Parametro di posizione (loc): {loc_hat}")
print(f"Parametro di scala (scale): {scale_hat}")

# Calcolo CDF
cdf = gamma_dist.cdf(x)

plt.figure()
plt.plot(x, cdf, 'b-', label='CDF stimata')
plt.title("Funzione di ripartizione (CDF) della Gamma(a=1)")
plt.xlabel("x")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Calcolo della varianza campionaria
var_sample = np.var(sample, ddof=1)
print(f"Varianza campionaria: {var_sample:.4f}")

```

### **2. Fitta i dati**
Prova a fittare i dati sottostante con le migliori curve, calcola il MAE e l'RMSE

```python
import numpy as np
temp_max = np.array([17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18])
temp_min = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])
import matplotlib.pyplot as plt
months = np.arange(12)
plt.plot(months, temp_max, "ro")
plt.plot(months, temp_min, "bo")
plt.xlabel("Month")
plt.ylabel("Min and max temperature")


# Fit polinomiale grado 3
coeffs_max = np.polyfit(months, temp_max, 3)
coeffs_min = np.polyfit(months, temp_min, 3)

# Predizioni
poly_max = np.poly1d(coeffs_max)
poly_min = np.poly1d(coeffs_min)
pred_max = poly_max(months)
pred_min = poly_min(months)

# Calcolo MAE e RMSE
mae_max = np.mean(np.abs(temp_max - pred_max))
rmse_max = np.sqrt(np.mean((temp_max - pred_max)**2))
mae_min = np.mean(np.abs(temp_min - pred_min))
rmse_min = np.sqrt(np.mean((temp_min - pred_min)**2))

# Plot
plt.figure(figsize=(8,6))
plt.plot(months, temp_max, "ro", alpha=0.5, label='Temp max (osservate)')
plt.plot(months, temp_min, "bo", alpha=0.5, label='Temp min (osservate)')
plt.plot(months, pred_max, color='r', linestyle='--', label='Fit max')
plt.plot(months, pred_min, color='b', linestyle='--', label='Fit min')
plt.title("Fitting polinomiale grado 3")
plt.xlabel("Month")
plt.ylabel("Min and max temperature")
plt.legend()
plt.show()

# Stampa dei risultati
print("Massime:")
print(f"  MAE = {mae_max:.2f}")
print(f"  RMSE = {rmse_max:.2f}")
print("Minime:")
print(f"  MAE = {mae_min:.2f}")
print(f"  RMSE = {rmse_min:.2f}")


# Fit polinomiale grado 5
coeffs_max = np.polyfit(months, temp_max, 5)
coeffs_min = np.polyfit(months, temp_min, 5)

# Predizioni
poly_max = np.poly1d(coeffs_max)
poly_min = np.poly1d(coeffs_min)
pred_max = poly_max(months)
pred_min = poly_min(months)

# Calcolo MAE e RMSE
mae_max = np.mean(np.abs(temp_max - pred_max))
rmse_max = np.sqrt(np.mean((temp_max - pred_max)**2))
mae_min = np.mean(np.abs(temp_min - pred_min))
rmse_min = np.sqrt(np.mean((temp_min - pred_min)**2))

# Plot
plt.figure(figsize=(8,6))
plt.plot(months, temp_max, "ro", alpha=0.5, label='Temp max (osservate)')
plt.plot(months, temp_min, "bo", alpha=0.5, label='Temp min (osservate)')
plt.plot(months, pred_max, color='r', linestyle='--', label='Fit max')
plt.plot(months, pred_min, color='b', linestyle='--', label='Fit min')
plt.title("Fitting polinomiale grado 5")
plt.xlabel("Month")
plt.ylabel("Min and max temperature")
plt.legend()
plt.show()

# Stampa dei risultati
print("Massime:")
print(f"  MAE = {mae_max:.2f}")
print(f"  RMSE = {rmse_max:.2f}")
print("Minime:")
print(f"  MAE = {mae_min:.2f}")
print(f"  RMSE = {rmse_min:.2f}")
```

### **3. Modello di regressione lineare dei seguenti dati**

 Calcola un modello di regressione lineare delle colonne mpg e disp del famoso dataset mtcars.

Dove: 

mpg = Miles Per Gallon → miglia per gallone, cioè una misura del consumo di carburante. Più alto è il valore, più efficiente è l’auto.

disp = Displacement → cilindrata del motore, in pollici cubici (cubic inches). Rappresenta il volume totale dei cilindri del motore. Più è alto, maggiore è la potenza potenziale del motore (ma anche il consumo).

```python
import pandas as pd

# Load dataset from URL
df = pd.read_csv('https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv')

# Display the dataframe
df
```

```python
# Estrazione colonne
mpg = df['mpg'].values
disp = df['disp'].values

# Coefficienti retta di regressione
slope, intercept = np.polyfit(disp, mpg, 1)

# Predizioni
mpg_pred = slope * disp + intercept

# R^2
ssr = np.sum((mpg - mpg_pred)**2)
sst = np.sum((mpg - np.mean(mpg))**2)
r2= 1 - ssr/sst

# Plot
plt.figure(figsize=(8,6))
plt.scatter(disp, mpg, label='Dati osservati')
plt.plot(disp, mpg_pred, 'r', label='Fit')
plt.title('Regressione lineare: mpg ~ disp')
plt.xlabel('Displacement (disp)')
plt.ylabel('Miles Per Gallon (mpg)')
plt.legend()
plt.show()

# Coefficienti
print(f"Intercept = {intercept:.4f}")
print(f"Slope     = {slope:.4f}")
print(f"R^2       = {r2:.4f}")
```
